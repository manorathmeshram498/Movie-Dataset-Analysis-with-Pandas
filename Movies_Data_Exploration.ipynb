```python
# Median of release year
median_year = df['release_year'].median()

# Mode of movie genres
genre_mode = df['genre'].mode()[0]    # In case of multiple modes, this takes the first one

print("Median Release Year:", median_year)
print("Most Common Genre (Mode):", genre_mode)
```

```python
High_rated_movies = df[df['rating']>7]
print(High_rated_movies)
```

```python
# Sort by box office in descending order and get top 10
top_10_earning_movies = df.sort_values(by='box_office', ascending=False).head(10) # for  selecting number of movies to show from top
top_earning_movies = df.sort_values(by='box_office' , ascending=False)

print(top_earning_movies)
```

```python

# Create the 'profitability' column

df['profitability'] = df['box_office']/100000000

# Display the first few rows
print(df[['title', 'box_office', 'profitability']].head())
```

```python
genre_stats = df.groupby("genre")[['rating', 'box_office']].mean()

print(genre_stats)
```

```python
import matplotlib.pyplot as plt


avg_rating_by_year = df.groupby('release_year')['rating' ].mean().reset_index()

# Sort by release year
avg_rating_by_year = avg_rating_by_year.sort_values('release_year')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_rating_by_year['release_year'], avg_rating_by_year['rating'], marker='o', linestyle='-')
plt.title('Average Movie Rating Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
